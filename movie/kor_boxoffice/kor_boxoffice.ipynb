```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

pd.set_option('display.max_columns', 100)
warnings.filterwarnings('ignore')

plt.rc('font', family= 'Malgun Gothic')
%matplotlib inline

df= pd.read_csv('kor_boxoffice.csv')
df.head()
```

```python
# 순번 컬럼 삭제
df.drop('순번', axis= 1, inplace= True)
df.head()
```

```python
# 데이터 전처리

# 개봉일 -> 날짜로
df['개봉일']= pd.to_datetime(df['개봉일'])

df.head()
```

```python
# 스크린수, 매출액, 관객수-> 쉼표 제거-> 정수
cols= ['전국 스크린수', '전국 매출액', '전국 관객수', '서울 매출액', '서울 관객수']

for col in cols :
    df[col]= df[col].apply(lambda x: x.replace(',', ''))
    df[col]= pd.to_numeric(df[col])

df.info()
```

```python
df.head()
```

```python
# 제작사 별 총 매출액
rev_rnk_credit= df['전국 매출액'].groupby(df['제작사']).sum().reset_index().sort_values(by= '전국 매출액', ascending= False).head(15)

plt.figure(figsize= (20, 10))
plt.xticks(rotation= '90')
sns.barplot(data= rev_rnk_credit, x= '제작사', y= '전국 매출액')
plt.show()
```
